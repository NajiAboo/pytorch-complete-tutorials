## Creating tensors

```python
#Scalar
scalar = torch.tensor(7)
print(scalar)
print(f'dimenstions : {scalar.ndim}')
print(f"Value : {scalar.item()}")
```

```python
# Vector
vector = torch.tensor([7,9])
print(vector)
print(f"dimension : {vector.ndim}")
```

```python
# Matrix
matrix = torch.tensor([[7, 10, 10], [8,8,9]])
print(matrix)
```

### Tensor Manipulation

Tensor Operations:

 * Subtraction
 * Multiplication (element-wise)
 * Division
 * Matrix Multiplication

```python
tensor_1 = torch.tensor([1, 2, 3])
tensor_1 + 10
```

```python
#multiply tensor
tensor_1 = torch.tensor([1, 2, 3])
tensor_1 * 10
```

```python
#subtract
tensor_1 - 10
```

```python
#Pytorch inbuilt
torch.mul(tensor_1, 10)
```
